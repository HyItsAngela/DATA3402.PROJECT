### Variable groups/features:

### Goal: 
assess the likelihood of the patient's diagnostic period being less than 90 days is predictable using these characteristics
### Target 
= DiagPeriodL90D

```python
target = 'DiagPeriodL90D'

# count the points of each class in target
t_counts = df[target].value_counts()

# plot class distribution
t_counts.plot(kind='bar')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# pie chart was easier to visualize percentage difference

t_counts.plot(kind='pie',figsize = (6, 6), autopct = '%1.1f%%', shadow = True)

plt.title('Class Distribution')
plt.ylabel('Count')
plt.legend(['1','0'])
plt.show()
```

#### Note:
Obeserve a binary classification. About a 4,000 point difference with 8:5 ratio and 63/38 percent difference. A class imbalance exists.

## 3 - Missing Values

### 3.1 - Training Data

```python
np.any(df.isnull())
```

```python
df.isnull().sum()
```

```python
# percentage of missing values in respective feature row
((df.isnull().sum()) / len(df)) * 100
```
